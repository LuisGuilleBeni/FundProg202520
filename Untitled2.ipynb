```python
# Juego de Mesa Simple - L√≠nea de Producci√≥n
print("=== JUEGO DE MESA - L√çNEA DE PRODUCCI√ìN ===")
print("El primero en llegar a la casilla 43 gana!")

# Preguntar cu√°ntos jugadores
while True:
    try:
        num_jugadores = int(input("¬øCu√°ntos jugadores? (m√≠nimo 2): "))
        if num_jugadores >= 2:
            break
        else:
            print("Tiene que ser al menos 2 jugadores")
    except:
        print("Escribe un n√∫mero v√°lido")

# Listas para guardar datos de jugadores
nombres = []
posiciones = []

# Pedir nombres de jugadores
for i in range(num_jugadores):
    nombre = input(f"Nombre del jugador {i+1}: ")
    nombres.append(nombre)
    posiciones.append(0)  # Todos empiezan en 0

print("\n¬°Que empiece el juego!")

# Variable para saber si hay ganador
ganador = None
ronda = 1

# Mientras no haya ganador, seguir jugando
while ganador is None:
    print(f"\n--- RONDA {ronda} ---")

    # Cada jugador juega su turno
    for i in range(num_jugadores):
        # Si ya hay ganador, terminar
        if ganador is not None:
            break

        print(f"\nTurno de {nombres[i]}")
        input("Presiona ENTER para lanzar el dado...")

        # Lanzar dado
        dado = random.randint(1, 6)
        print(f"üé≤ Sacaste: {dado}")

        # Mover jugador
        posiciones[i] += dado
        print(f"Est√°s en la casilla: {posiciones[i]}")

        # Verificar si cay√≥ en casilla especial
        casilla_actual = posiciones[i]

        # Casilla 5: Ajuste de torque
        if casilla_actual == 5:
            print("¬°Ajuste de torque! Avanzas 2 casillas m√°s")
            posiciones[i] += 2
            print(f"Nueva posici√≥n: {posiciones[i]}")

        # Casilla 10: Sensor fuera de calibraci√≥n
        elif casilla_actual == 10:
            print("¬°Sensor descalibrado! Vas a la casilla 12")
            posiciones[i] = 12
            print(f"Nueva posici√≥n: {posiciones[i]}")

        # Casilla 15: Modo autom√°tico
        elif casilla_actual == 15:
            print("¬°Modo autom√°tico! Juegas otra vez")
            # Turno extra
            print("--- TURNO EXTRA ---")
            dado_extra = random.randint(1, 6)
            print(f"üé≤ Sacaste: {dado_extra}")
            posiciones[i] += dado_extra
            print(f"Nueva posici√≥n: {posiciones[i]}")

        # Casilla 20: Mantenimiento preventivo
        elif casilla_actual == 20:
            print("¬°Mantenimiento! Todos retroceden 1 casilla")
            for j in range(num_jugadores):
                if posiciones[j] > 0:
                    posiciones[j] -= 1
            print("Posiciones actualizadas:")
            for j in range(num_jugadores):
                print(f"  {nombres[j]}: {posiciones[j]}")

        # Casilla 25: Actualizaci√≥n de software
        elif casilla_actual == 25:
            print("¬°Actualizaci√≥n! Vas a la casilla 28")
            posiciones[i] = 28
            print(f"Nueva posici√≥n: {posiciones[i]}")

        # Casilla 30: Optimizaci√≥n del sistema
        elif casilla_actual == 30:
            print("¬°Optimizaci√≥n! Elige:")
            print("1. Avanzar 3 casillas")
            print("2. Retroceder 2 casillas")

            while True:
                try:
                    opcion = int(input("Tu elecci√≥n (1 o 2): "))
                    if opcion == 1:
                        posiciones[i] += 3
                        print("Avanzas 3 casillas")
                        break
                    elif opcion == 2:
                        posiciones[i] -= 2
                        if posiciones[i] < 0:
                            posiciones[i] = 0
                        print("Retrocedes 2 casillas")
                        break
                    else:
                        print("Escribe 1 o 2")
                except:
                    print("Escribe un n√∫mero v√°lido")
            print(f"Nueva posici√≥n: {posiciones[i]}")

        # Casilla 35: Falla mec√°nica
        elif casilla_actual == 35:
            print("¬°Falla mec√°nica! Retrocedes a la casilla 23")
            posiciones[i] = 23
            print(f"Nueva posici√≥n: {posiciones[i]}")

        # Casilla 40: Restablecimiento de energ√≠a
        elif casilla_actual == 40:
            print("¬°Restablecimiento de energ√≠a!")
            print("T√∫ avanzas 6 casillas, los dem√°s retroceden 6")
            posiciones[i] += 6

            for j in range(num_jugadores):
                if j != i:  # Para los otros jugadores
                    posiciones[j] -= 6
                    if posiciones[j] < 0:
                        posiciones[j] = 0

            print("Posiciones actualizadas:")
            for j in range(num_jugadores):
                print(f"  {nombres[j]}: {posiciones[j]}")

        # Casilla 50: L√≠nea de producci√≥n - GANAR
        elif casilla_actual >= 43:
            ganador = i
            print(f"üéâ ¬°{nombres[i]} lleg√≥ a la l√≠nea de producci√≥n!")
            print("¬°¬°¬°GANASTE EL JUEGO!!!")
            break

    # Mostrar posiciones despu√©s de cada ronda
    print(f"\n--- RESUMEN RONDA {ronda} ---")
    for j in range(num_jugadores):
        print(f"{nombres[j]}: casilla {posiciones[j]}")

    ronda += 1

    # Para no hacer demasiadas rondas
    if ronda > 43:
        print("\nDemasiadas rondas, el juego termina")
        # Buscar quien est√° m√°s adelante
        max_posicion = 0
        for j in range(num_jugadores):
            if posiciones[j] > max_posicion:
                max_posicion = posiciones[j]
                ganador = j
        break

# Mostrar resultado final
print(f"\n{'='*33}")
if ganador is not None:
    print(f"üèÜ ¬°EL GANADOR ES {nombres[ganador].upper()}! üèÜ")
else:
    print("El juego termin√≥ sin ganador")
print(f"{'='*33}")

print("\n¬°Gracias por jugar!")
```
